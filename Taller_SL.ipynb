Considere la ecuación diferencial $\frac{d^2\varphi}{dx^2}+k^2\varphi=0$, las cual tiene la solución general $\varphi(x)=Asen(kx)+Bcos(kx)$. Si esta ed esta sujeta a las condiciones de frontera $\varphi(0)=\varphi(L)=0$, la solución general y sus autovalores admiten la siguiente escritura:
$$\varphi_n(x)=A_nsen(n\pi x/L),\hspace{1cm}n=1,2,3,...$$
$$\lambda_n=k^2_n=(n\pi/L)^2.$$

```python
#tamaño caja
L=1

#dominio 
x = np.linspace(0,L,100)

# función y autovalores

def φ(x,n):
    return np.sin(n*np.pi*x/L)

def k(n):
    return (n*np.pi/L)
```

```python
n = 3 #número autofunciones
I = np.zeros((n,n))

for i in range(n):
    for j in range(n):
        if i!=j:
            I[i][j] = np.sin((i+j)*np.pi)/((i+j)*np.pi)-np.sin((j-i)*np.pi)/((j-i)*np.pi)
        if i==j:
            I[i][j] = 1
print(I)
```

Considere la ecuación diferencial $x^2\ddot{y}+5x\dot{y}+\lambda y=0$, para $1<x<e$, con lo cual $y(1)=y(e)=0$. La solución general y sus autovalores admiten la siguiente escritura:
$$y_n(x)=\sqrt{\frac{2}{\pi}}\frac{sen(n\pi ln(x))}{x^2},\hspace{1cm}n=1,2,3,...$$
$$\lambda_n=4+n^2{\pi}^2.$$

```python
#dominio 
x = np.linspace(1,np.e,100)

# función y autovalores

def y(x,n):
    return np.sqrt(2/np.pi)*np.sin(n*np.pi*np.log(x))/(x**2)

def λ(n):
    return 4+n**2*np.pi*np.pi
```

```python
#grafica
plt.figure(figsize=(15,7))

#lineas verticales
plt.axvline(1, 0, 1, label='pyplot vertical line',color='k')
plt.axvline(np.e, 0, 1, label='pyplot vertical line',color='k')

#autofunciones
for i in range(5):
    #plt.plot(x,y(x,i),lw=1)
    plt.plot(x,y(x,i)+np.sqrt(λ(i)),lw=1)


plt.rc('xtick',color='w')
plt.rc('ytick',color='w')
#plt.legend(loc="upper left")
```

$$\sum_{n=1}^{\infty}\frac{\cos(nx)}{n}=-\ln\left(2\sin\left(\frac{x}{2}\right)\right)$$

$0<x<2\pi$

Función de Bessel:

$$J_n(x)=\sum_{k=0}^{\infty}\frac{(-1)^k}{k!(k+n)!}(\frac{x}{2})^{2k+n}$$

```python
x = np.linspace(0,40,200)
N = 120 # número funciones
```

```python
xx = len(x)
J = np.zeros((10,xx))

for i in range(10):
    for j in range(xx):
        J[i][j] = sum([ (-1)**(k)/((math.factorial(k)*(math.factorial(k+i))))*(x[j]/2)**(2*k+i) for k in range(N)])
        
```

Polinomios de Legendre:

$$P_l(x)=\sum_{k=0}^{\infty}\frac{(-1)^k(2l-2k)!}{2^lk!(l-k)!(l-2k)!}x^{l-2k}$$

```python
# Valor intervalo
L = 1.5

#dominio 
x = np.linspace(-L,L,400)
#número de funciones
N = 700
# número l 
LL = 45
```

```python
xx = len(x)
P = np.zeros((LL,xx))

for l in range(LL):
    for j in range(xx):
        
        suma = 0
        
        for k in range(N):
            
            if (2*l-2*k) <0:
                continue
            if (l-k) < 0:
                continue
            if  (l-2*k) < 0:
                continue
                
            else: 
                suma += (x[j])**(l-2*k)*((-1)**k*math.factorial(2*l-2*k)/(2**l
                        *math.factorial(k)*math.factorial(l-k)
                         *math.factorial(l-2*k)))              
        P[l][j] = suma
        
```

```python
plt.figure(figsize=(15,7))

for a in range(7):
    plt.plot(x,P[a],lw=2,linestyle='-')

    
plt.xlim(-L,L)
plt.ylim(-L,L+0.03)


plt.rc('xtick',color='w')
plt.rc('ytick',color='w')
```
